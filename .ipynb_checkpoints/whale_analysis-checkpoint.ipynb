## Whale Returns

Read the Whale Portfolio daily returns and clean the data

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_returns_csv = Path("Resources/algo_returns.csv")
# YOUR CODE HERE
algo_df = pd.read_csv(algo_returns_csv, index_col='Date', infer_datetime_format=True, parse_dates=True)
algo_df.head()
```

```python
# Count nulls
# YOUR CODE HERE
algo_df.isnull().sum()
```

```python
# Drop nulls
# YOUR CODE HERE
algo_df.dropna(inplace=True)
algo_df.isnull().sum()
```

## S&P 500 Returns

Read the S&P500 Historic Closing Prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices, sorting index
sp500_history_csv = Path("Resources/sp500_history.csv")
# YOUR CODE HERE
sp500_df = pd.read_csv(sp500_history_csv, index_col='Date', infer_datetime_format=True, parse_dates=True)
sp500_df.head()
```

```python
# Check Data Types
# YOUR CODE HERE
sp500_df.dtypes
```

```python
# Fix Data Types
# YOUR CODE HERE
sp500_df['Close'] = sp500_df['Close'].str.replace('$', '')
sp500_df['Close'] = sp500_df['Close'].astype(float)
sp500_df.head()
```

```python
# Calculate Daily Returns
# YOUR CODE HERE
daily_returns_sp500 = sp500_df.pct_change()
daily_returns_sp500
```

```python
# Drop nulls
# YOUR CODE HERE
daily_returns_sp500.isnull().sum()
```

```python
daily_returns_sp500.dropna(inplace=True)
daily_returns_sp500.isnull().sum()
```

```python
daily_returns_sp500.head()
```

```python
# Rename Column
# YOUR CODE HERE
daily_returns_sp500.columns = ['SP500']
daily_returns_sp500.head()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Concatenate all DataFrames into a single DataFrame
# YOUR CODE HERE
combined_df = pd.concat([whale_df, algo_df, daily_returns_sp500], axis='columns', join='inner')
combined_df.sort_index(inplace=True)
combined_df.head()
```

Calculate and Plot the daily returns and cumulative returns.

Determine the _risk_ of each portfolio:

1. Create a box plot for each portfolio. 
2. Calculate the standard deviation for all portfolios
4. Determine which portfolios are riskier than the S&P 500

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for the S&PP 500 using a 21 day window
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P 500
2. Calculate and plot a 60 day Beta for Berkshire Hathaway Inc compared to the S&&P 500

```python
# Calculate and plot the rolling standard deviation for the S&PP 500 using a 21 day window
# YOUR CODE HERE
combined_df.rolling(window=21).std().plot(figsize=(20,10))
```

**Risk does increase same time as SP500 for portfolios, but the risk amount does not correlate**

```python
# Correlation
# YOUR CODE HERE
correlation = combined_df.corr()
correlation
```

**Seems like none of the portfolios very closely correlate with SP500, closest would be Paulson&CO (closest to perfect correlation of 1)**

```python
# Calculate Beta for a single portfolio compared to the total market (S&P 500)
# YOUR CODE HERE
covariance = combined_df['BERKSHIRE HATHAWAY INC'].rolling(window=21).cov(combined_df['SP500'])
covariance

variance = combined_df['BERKSHIRE HATHAWAY INC'].rolling(window=21).var()
variance

berkshire_beta = covariance / variance
berkshire_beta.plot(figsize=(10,5))
```

**Portfolio does seem sensitive as it moves quite a bit up/down compared to SP500 movements**

## Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. (After all, if you could invest in one of two portfolios, each offered the same 10% return, yet one offered lower risk, you'd take that one, right?)

Calculate and plot the annualized Sharpe ratios for all portfolios to determine which portfolio has the best performance

```python
# Annualzied Sharpe Ratios
# YOUR CODE HERE
# Use the `mean` and `std` functions to calculate the annualized sharpe ratio
sharpe_ratios = (combined_df.mean() * 252) / (combined_df.std() * np.sqrt(252))
sharpe_ratios
```

 plot() these sharpe ratios using a barplot.
 On the basis of this performance metric, do our algo strategies outperform both 'the market' and the whales?

```python
# Visualize the sharpe ratios as a bar plot
# YOUR CODE HERE
sharpe_ratios.plot(kind='bar')
```

**Algo1 outperforms all, algo2 is being outperformed by Berkshire (and algo1 obviously)**

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

```python
# Read the first stock
# YOUR CODE HERE
advance_path=Path('Resources', 'ADVANCE.csv')
advance_df=pd.read_csv(advance_path, index_col='Date', parse_dates=True)
advance_df.index = advance_df.index.normalize()
advance_df.columns = ['Advance']
advance_df.sort_index(ascending=False, inplace=True)
advance_df.head()
```

```python
# Read the second stock
# YOUR CODE HERE
blizzard_path=Path('Resources', 'BLIZZARD.csv')
blizzard_df=pd.read_csv(blizzard_path, index_col='Date', parse_dates=True)
blizzard_df.index = blizzard_df.index.normalize()
blizzard_df.columns = ['Blizzard']
blizzard_df.sort_index(ascending=False, inplace=True)
blizzard_df.head()
```

```python
# Read the third stock
# YOUR CODE HERE
snapon_path=Path('Resources', 'SNAP.csv')
snapon_df=pd.read_csv(snapon_path, index_col='Date', parse_dates=True)
snapon_df.index = snapon_df.index.normalize()
snapon_df.columns = ['SnapOn']
snapon_df.sort_index(ascending=False, inplace=True)
snapon_df.head()
```

```python
# Concatenate all stocks into a single DataFrame
# YOUR CODE HERE
combined_stocks_df = pd.concat([advance_df, blizzard_df, snapon_df], axis='columns', join='inner')
combined_stocks_df.head()
```

```python
# Reset the index
# YOUR CODE HERE
combined_stocks_df.reset_index(drop=True)
combined_stocks_df.sort_index(inplace=True)
combined_stocks_df.tail()
```

```python
# Drop Nulls
# YOUR CODE HERE
combined_stocks_df.isnull().sum()

#NO NULLS
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Calculate weighted portfolio returns
weights = [1/3, 1/3, 1/3]
# YOUR CODE HERE
daily_stock_returns_df = combined_stocks_df.pct_change()
portfolio_returns_df = daily_stock_returns_df.dot(weights)
portfolio_returns_df.sort_index(inplace=True)
portfolio_returns_df.tail()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# YOUR CODE HERE
combined_portfolios_df = pd.concat([portfolio_returns_df, combined_df], axis='columns', join='inner')
combined_portfolios_df.columns.values[0] = 'DKV'
combined_portfolios_df.tail()
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
# YOUR CODE HERE
combined_portfolios_df.isnull().sum()
combined_portfolios_df = combined_portfolios_df.dropna()
combined_portfolios_df.tail()
```

## Re-run the performance and risk analysis with your portfolio to see how it compares to the others

```python
# Risk
# YOUR CODE HERE
combined_portfolios_df.plot.box(figsize=(25,10))
```

**Quite the spread for DKV in comparison! Logical if you think that DKV portfolio consists of a mere 3 stocks whereas the others are mixed funds/portfolios consisting of many more underlying assets.**

```python
daily_std_combined_portfolios = combined_portfolios_df.std()
daily_std_combined_portfolios = daily_std_combined_portfolios.sort_values(ascending=False)
daily_std_combined_portfolios
```

```python
annualized_std_portfolios = daily_std_combined_portfolios * np.sqrt(252)
annualized_std_portfolios
```

**Risky DKV! great that I have a big risk-appetite!**

```python
# Rolling
# YOUR CODE HERE
combined_portfolios_df.rolling(window=21).std().plot(figsize=(20,10))
```

```python
# Beta
# YOUR CODE HERE
covariance_DKV = combined_portfolios_df['DKV'].rolling(window=21).cov(combined_portfolios_df['SP500'])
covariance_DKV

variance_DKV = combined_portfolios_df['DKV'].rolling(window=21).var()
variance_DKV

beta_DKV = covariance_DKV / variance_DKV
beta_DKV.plot(figsize=(10,5))
```

**DKV also quite sensitive to SP500 movements, confirming volatility and risk**

```python
# Annualzied Sharpe Ratios
# YOUR CODE HERE
sharpe_ratios_portfolios = (combined_portfolios_df.mean() * 252) / (combined_portfolios_df.std() * np.sqrt(252))
sharpe_ratios_portfolios
```

```python
# Visualize the sharpe ratios as a bar plot
# YOUR CODE HERE
sharpe_ratios_portfolios.plot(kind='bar')
```

**not bad DKV, outperforming Paulson, Tiger and SP500!**
